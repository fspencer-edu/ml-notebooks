## Imports

````
from transformers import pipeline, AutoTokenizer, AutoModel, AutoModelForSequenceClassification
import torch
print("Transformers imported successfully.")
````

## Pipeline usage

````
classifier = pipeline("sentiment-analysis")
print(classifier("I love studying machine learning."))
````

## Tokenizer

````
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
encoded = tokenizer("Hello from Hugging Face Transformers!", return_tensors="pt")
print(encoded)
````

## Model forward pass

````
model = AutoModel.from_pretrained("distilbert-base-uncased")
with torch.no_grad():
    outputs = model(**encoded)
print(outputs.last_hidden_state.shape)
````

## Sequence classification model

````
clf_model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
print(clf_model)
````
