## Imports and downloads

````python
import nltk
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
````

## Tokenization

````python
text = "NLTK is useful for learning natural language processing. It has many classic tools."
print(sent_tokenize(text))
print(word_tokenize(text))
````

## Stopwords, stemming, lemmatization

````python
tokens = word_tokenize(text.lower())
sw = set(stopwords.words("english"))
filtered = [t for t in tokens if t.isalpha() and t not in sw]
print("Filtered:", filtered)
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
print("Stemmed:", [stemmer.stem(t) for t in filtered])
print("Lemmatized:", [lemmatizer.lemmatize(t) for t in filtered])
````
