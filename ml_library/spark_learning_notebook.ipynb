## Spark session

````
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
spark = SparkSession.builder.appName("spark-demo").getOrCreate()
print(spark)
````

## Create DataFrame

````
data = [("Fiona", "Toronto", 92), ("Alex", "Montreal", 85), ("Sam", "Toronto", 78)]
df = spark.createDataFrame(data, ["name", "city", "score"])
df.show()
````

## Select, filter, group

````
df.select("name", "score").show()
df.filter(F.col("score") >= 85).show()
df.groupBy("city").agg(F.avg("score").alias("avg_score")).show()
````
