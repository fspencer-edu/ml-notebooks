## Imports

````
from fastai.tabular.all import *
from sklearn.datasets import load_iris
import pandas as pd
print("fastai imported successfully.")
````

## Tabular workflow

````
iris = load_iris(as_frame=True)
df = iris.frame
df["target"] = df["target"].map(dict(enumerate(iris.target_names)))
df.head()
````

## DataLoaders and Learner

````
dls = TabularDataLoaders.from_df(
    df, path='.', y_names='target', cat_names=[],
    cont_names=iris.feature_names, procs=[Categorify, FillMissing, Normalize], bs=16
)
learn = tabular_learner(dls, metrics=accuracy)
learn.fit(5)
````

## Prediction and interpretation

````
row = df.iloc[0]
pred, pred_idx, probs = learn.predict(row)
print("Prediction:", pred)
print("Probabilities:", probs)
````
