## Basic environment loop

````python
import gymnasium as gym

env = gym.make("CartPole-v1")
obs, info = env.reset()

for step in range(5):
    action = env.action_space.sample()
    obs, reward, terminated, truncated, info = env.step(action)
    print("step:", step, "reward:", reward, "done:", terminated or truncated)
    if terminated or truncated:
        obs, info = env.reset()

env.close()
````

## Useful pieces

````python
reference = {
    "Core": ["make", "reset", "step", "close"],
    "Spaces": ["action_space", "observation_space", "sample"]
}
for k, v in reference.items():
    print("\n" + k + ":")
    print(", ".join(v))
````
