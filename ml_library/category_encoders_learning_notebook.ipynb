## Example data

````
import pandas as pd
from category_encoders import OneHotEncoder, TargetEncoder, OrdinalEncoder

X = pd.DataFrame({
    "city": ["Toronto", "Montreal", "Toronto", "Vancouver", "Montreal"],
    "plan": ["A", "A", "B", "B", "C"]
})
y = pd.Series([1, 0, 1, 0, 0])
X
````

## One-hot encoding

````
enc = OneHotEncoder(use_cat_names=True)
X_ohe = enc.fit_transform(X)
X_ohe
````

## Target encoding

````
tenc = TargetEncoder()
X_te = tenc.fit_transform(X, y)
X_te
````
