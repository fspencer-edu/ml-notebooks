## EntitySet and DFS

````python
import pandas as pd
import featuretools as ft

customers = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "join_date": pd.to_datetime(["2025-01-01", "2025-01-15", "2025-02-01"])
})

transactions = pd.DataFrame({
    "transaction_id": [101, 102, 103, 104],
    "customer_id": [1, 1, 2, 3],
    "amount": [100, 35, 50, 80],
    "transaction_time": pd.to_datetime(["2025-02-01", "2025-02-02", "2025-02-03", "2025-02-04"])
})

es = ft.EntitySet(id="demo")
es = es.add_dataframe(dataframe_name="customers", dataframe=customers, index="customer_id", time_index="join_date")
es = es.add_dataframe(dataframe_name="transactions", dataframe=transactions, index="transaction_id", time_index="transaction_time")
es = es.add_relationship("customers", "customer_id", "transactions", "customer_id")

feature_matrix, feature_defs = ft.dfs(
    entityset=es,
    target_dataframe_name="customers",
    agg_primitives=["sum", "mean", "count"]
)
feature_matrix
````
