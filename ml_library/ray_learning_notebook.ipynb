## Init and remote functions

````
import ray
ray.init(ignore_reinit_error=True)

@ray.remote
def square(x):
    return x * x

futures = [square.remote(i) for i in range(5)]
print(ray.get(futures))
````

## Actors

````
@ray.remote
class Counter:
    def __init__(self):
        self.value = 0
    def inc(self):
        self.value += 1
        return self.value

c = Counter.remote()
print(ray.get(c.inc.remote()))
print(ray.get(c.inc.remote()))
````
