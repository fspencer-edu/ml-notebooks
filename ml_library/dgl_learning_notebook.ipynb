## Create graph

````
import dgl
import torch

g = dgl.graph(([0, 1, 2], [1, 2, 3]))
g.ndata["feat"] = torch.randn(4, 3)
print(g)
print(g.ndata["feat"])
````

## Useful pieces

````
reference = {
    "Core": ["graph", "ndata", "edata"],
    "Graph ops": ["add_nodes", "add_edges", "number_of_nodes", "number_of_edges"],
    "NN": ["GraphConv", "GATConv", "SAGEConv"]
}
for k, v in reference.items():
    print("\n" + k + ":")
    print(", ".join(v))
````
