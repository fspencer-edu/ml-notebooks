## Imports

````
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
````

## Basic plot

````
p = figure(title="Bokeh Line Plot", width=700, height=400)
p.line([1,2,3,4], [3,7,5,8], line_width=2)
show(p)
````

## Useful pieces

````
reference = {
    "Plotting": ["figure", "line", "circle", "vbar", "show"],
    "Notebook": ["output_notebook"],
    "Customization": ["title", "legend_label", "width", "height"]
}
for k, v in reference.items():
    print("\n" + k + ":")
    print(", ".join(v))
````
