# DVC Learning Notebook


This notebook shows how to use **DVC** for data and pipeline versioning.

DVC is mainly a **CLI tool**, so this notebook focuses on the most useful commands and workflow.

## Jupyter setup

```bash
pip install dvc jupyter
jupyter notebook
```

## Common commands

````python
commands = [
    "dvc init",
    "dvc add data/train.csv",
    "git add data/train.csv.dvc .gitignore",
    "git commit -m 'Track dataset with DVC'",
    "dvc remote add -d storage s3://my-bucket/path",
    "dvc push",
    "dvc pull",
    "dvc repro"
]
for cmd in commands:
    print(cmd)
````

## Pipeline example

````python
pipeline = '''
stages:
  train:
    cmd: python train.py
    deps:
      - train.py
      - data/train.csv
    outs:
      - models/model.pkl
'''
print(pipeline)
````
