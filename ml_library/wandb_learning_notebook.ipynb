## Basic usage

````
import wandb

# Run wandb.login() once in your environment before tracking online.
# wandb.login()

run = wandb.init(project="demo-project", mode="disabled")
for epoch in range(5):
    wandb.log({"epoch": epoch, "loss": 1/(epoch+1), "accuracy": 0.7 + epoch*0.05})
run.finish()
print("W&B run finished.")
````

## Useful pieces

````
reference = {
    "Core": ["init", "log", "finish", "login"],
    "Common data": ["metrics", "config", "artifacts", "tables"]
}
for k, v in reference.items():
    print("\n" + k + ":")
    print(", ".join(v))
````
