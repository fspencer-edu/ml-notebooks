## Imports

````python
import altair as alt
from vega_datasets import data
df = data.cars()
df.head()
````

## Basic chart

````python
chart = alt.Chart(df).mark_circle(size=60).encode(
    x="Horsepower",
    y="Miles_per_Gallon",
    color="Origin",
    tooltip=["Name", "Horsepower", "Miles_per_Gallon"]
).interactive()
chart
````

## Useful pieces

````python
reference = {
    "Core": ["Chart", "mark_line", "mark_bar", "mark_circle", "encode"],
    "Encodings": ["x", "y", "color", "tooltip", "size"],
    "Options": ["interactive", "properties", "transform_filter"]
}
for k, v in reference.items():
    print("\n" + k + ":")
    print(", ".join(v))
````
