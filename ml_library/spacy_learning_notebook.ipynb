## Imports and model

````python
import spacy
nlp = spacy.load("en_core_web_sm")
print("spaCy version:", spacy.__version__)
````

## Doc, tokens, sentences

````python
doc = nlp("Fiona studies machine learning in Toronto. She likes Python and NLP.")
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.dep_)
print("\nSentences:")
for sent in doc.sents:
    print("-", sent.text)
````

## Named entities

````python
for ent in doc.ents:
    print(ent.text, ent.label_)
````
