## Imports

````python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
````

## Plotly Express

````python
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", title="Interactive Scatter")
fig.show()
````

## Graph Objects

````python
fig = go.Figure()
fig.add_trace(go.Bar(x=["A", "B", "C"], y=[10, 20, 15], name="Sales"))
fig.update_layout(title="Bar Chart", xaxis_title="Category", yaxis_title="Value")
fig.show()
````
