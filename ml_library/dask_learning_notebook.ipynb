## Dask DataFrame

````python
import dask.dataframe as dd
import pandas as pd
pdf = pd.DataFrame({"x": range(100), "y": range(100, 200)})
ddf = dd.from_pandas(pdf, npartitions=4)
print(ddf)
print(ddf.head())
````

## Computation

````python
print(ddf["y"].mean().compute())
print(ddf.groupby(ddf["x"] % 2)["y"].mean().compute())
````

## Delayed

````python
from dask import delayed

@delayed
def square(x):
    return x * x

tasks = [square(i) for i in range(5)]
print(tasks)
print(delayed(sum)(tasks).compute())
````
