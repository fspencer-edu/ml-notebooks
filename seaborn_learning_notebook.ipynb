````python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

print("Seaborn version:", sns.__version__)
print("Pandas version:", pd.__version__)
````

## 2) Load simple datasets

````python

# Built-in Seaborn datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
penguins = sns.load_dataset("penguins")

print("tips shape:", tips.shape)
print("iris shape:", iris.shape)
print("penguins shape:", penguins.shape)

tips.head()
````

## 3) Themes and style

````python

sns.set_theme(style="darkgrid")
print("Current theme set to darkgrid")
````

````python

# Try a few common styles
for style in ["darkgrid", "whitegrid", "dark", "white", "ticks"]:
    sns.set_theme(style=style)
    plt.figure(figsize=(6, 3))
    sns.lineplot(data=iris, x=iris.index, y="sepal_length")
    plt.title(f"Style: {style}")
    plt.show()

# Reset to a common default
sns.set_theme(style="darkgrid")
````

## 4) Color palettes

````python

print("Some built-in palettes:")
print(["deep", "muted", "bright", "pastel", "dark", "colorblind"])

sns.palplot(sns.color_palette("deep"))
plt.title("Palette: deep")
plt.show()
````

## 5) Line plots

````python

plt.figure(figsize=(8, 5))
sns.lineplot(data=iris, x=iris.index, y="sepal_length")
plt.title("Line Plot")
plt.show()
````

````python

plt.figure(figsize=(8, 5))
sns.lineplot(data=iris, x=iris.index, y="petal_length", hue="species")
plt.title("Line Plot with hue")
plt.show()
````

## 6) Scatter plots

````python

plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species", style="species")
plt.title("Scatter Plot")
plt.show()
````

## 7) Relational figure-level plots

````python

sns.relplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species",
    col="species",
    kind="scatter",
    height=4,
    aspect=1
)
plt.show()
````

## 8) Histograms

````python

plt.figure(figsize=(8, 5))
sns.histplot(data=iris, x="sepal_length", bins=15)
plt.title("Histogram")
plt.show()
````

````python

plt.figure(figsize=(8, 5))
sns.histplot(data=iris, x="sepal_length", hue="species", bins=15, kde=True)
plt.title("Histogram with hue and KDE")
plt.show()
````

## 9) KDE plots

````python

plt.figure(figsize=(8, 5))
sns.kdeplot(data=iris, x="sepal_length", fill=True)
plt.title("KDE Plot")
plt.show()
````

````python

plt.figure(figsize=(8, 5))
sns.kdeplot(data=iris, x="sepal_length", y="petal_length", fill=True, cmap="Blues")
plt.title("2D KDE Plot")
plt.show()
````

## 10) ECDF plots

````python

plt.figure(figsize=(8, 5))
sns.ecdfplot(data=iris, x="sepal_length", hue="species")
plt.title("ECDF Plot")
plt.show()
````

## 11) Rug plots

````python

plt.figure(figsize=(8, 5))
sns.rugplot(data=iris, x="sepal_length")
plt.title("Rug Plot")
plt.show()
````

## 12) Bar plots

````python

plt.figure(figsize=(8, 5))
sns.barplot(data=tips, x="day", y="total_bill")
plt.title("Bar Plot")
plt.show()
````

````python

plt.figure(figsize=(8, 5))
sns.barplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Bar Plot with hue")
plt.show()
````

## 13) Count plots

````python

plt.figure(figsize=(8, 5))
sns.countplot(data=tips, x="day")
plt.title("Count Plot")
plt.show()
````

## 14) Box plots

````python

plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Box Plot")
plt.show()
````

````python

plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker")
plt.title("Box Plot with hue")
plt.show()
````

## 15) Violin plots

````python

plt.figure(figsize=(8, 5))
sns.violinplot(data=tips, x="day", y="total_bill")
plt.title("Violin Plot")
plt.show()
````

## 16) Strip and swarm plots

````python

plt.figure(figsize=(8, 5))
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Strip Plot")
plt.show()
````

````python

plt.figure(figsize=(8, 5))
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, size=4)
plt.title("Swarm Plot")
plt.show()
````

## 17) Point plots

````python

plt.figure(figsize=(8, 5))
sns.pointplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Point Plot")
plt.show()
````

## 18) Categorical figure-level plots

````python

sns.catplot(data=tips, x="day", y="total_bill", hue="sex", kind="box", height=4, aspect=1.3)
plt.show()
````

## 19) Regression plots

````python

plt.figure(figsize=(8, 5))
sns.regplot(data=tips, x="total_bill", y="tip")
plt.title("Regression Plot")
plt.show()
````

````python

sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex", height=5, aspect=1.2)
plt.show()
````

## 20) Pair plots

````python

sns.pairplot(iris, hue="species")
plt.show()
````

## 21) Joint plots

````python

sns.jointplot(data=iris, x="sepal_length", y="petal_length", kind="scatter", height=6)
plt.show()
````

````python

sns.jointplot(data=iris, x="sepal_length", y="petal_length", kind="kde", fill=True, height=6)
plt.show()
````

## 22) Heatmaps

````python

corr = iris.select_dtypes(include=np.number).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
````

## 23) Clustermap

````python

sns.clustermap(corr, annot=True, cmap="viridis")
plt.show()
````

## 24) FacetGrid

````python

g = sns.FacetGrid(tips, col="time", row="sex", height=3)
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.add_legend()
plt.show()
````

## 25) PairGrid

````python

g = sns.PairGrid(iris, hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.show()
````

## 26) Working with pandas DataFrames

````python

sales_df = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "sales": [120, 140, 150, 170, 165, 180],
    "profit": [20, 25, 23, 30, 28, 35],
    "region": ["East", "West", "East", "West", "East", "West"]
})

plt.figure(figsize=(8, 5))
sns.barplot(data=sales_df, x="month", y="sales", hue="region")
plt.title("Seaborn with pandas DataFrame")
plt.show()

sales_df
````

## 27) Common customization options

````python

plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species",
    style="species",
    size="petal_width",
    sizes=(20, 200)
)
plt.title("Customization: hue, style, size")
plt.grid(True)
plt.show()
````

## 28) Figure size, titles, labels, and legend placement

````python

plt.figure(figsize=(9, 5))
ax = sns.lineplot(data=sales_df, x="month", y="sales", hue="region", marker="o")
ax.set_title("Monthly Sales by Region")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")
plt.legend(title="Region", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
````

## 29) Useful Seaborn functions by category

````python

seaborn_reference = {
    "Relational": ["lineplot", "scatterplot", "relplot"],
    "Distribution": ["histplot", "kdeplot", "ecdfplot", "rugplot", "displot"],
    "Categorical": ["barplot", "countplot", "boxplot", "violinplot", "stripplot", "swarmplot", "pointplot", "catplot"],
    "Regression": ["regplot", "lmplot", "residplot"],
    "Matrix": ["heatmap", "clustermap"],
    "Multi-plot grids": ["FacetGrid", "PairGrid", "pairplot", "jointplot"],
    "Themes/palettes": ["set_theme", "set_style", "set_context", "color_palette", "palplot"],
    "Utilities": ["load_dataset", "despine", "move_legend"]
}

for category, funcs in seaborn_reference.items():
    print(f"\n{category}:")
    print(", ".join(funcs))
````

## 31) Figure-level vs axes-level functions

````python

axes_level = [
    "lineplot", "scatterplot", "histplot", "kdeplot", "barplot", "boxplot",
    "violinplot", "stripplot", "swarmplot", "pointplot", "regplot", "heatmap"
]

figure_level = [
    "relplot", "displot", "catplot", "lmplot", "pairplot", "jointplot", "clustermap"
]

print("Axes-level functions:")
for item in axes_level:
    print("-", item)

print("\nFigure-level functions:")
for item in figure_level:
    print("-", item)
````
