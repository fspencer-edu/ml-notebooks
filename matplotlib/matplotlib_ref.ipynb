# matplotlib Referece

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Basic plotting with pyplot
x = np.arange(0, 10)
y = x ** 2

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("x")
```

```python
# Fiture and Axes Objects
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y)
ax.set_title("Sine Wave")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
plt.show()

print("Figure object:", fig)
print("Axes object:", ax)
```

```python
# Common line plit
x = np.linspace(0, 10, 50)

plt.figure(figsize=(8, 5))
plt.plot(x, x, label="linear", linestyle="-", marker="o")
plt.plot(x, x**2, label="quadratic", linestyle="--", marker="s")
plt.plot(x, x**3, label="cubic", linestyle=":", marker="^")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter plots
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha=0.7)
plt.title("Scatter Plot")
plt.xlabel("x")
plt.ylabel("x")
plt.show()
```

```python
# Bar charts
categories = ["A", "B", "C", "D"]
values = [10, 25, 15, 30]

plt.figure(figsize=(8, 5))
plt.bar(categories, values)
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
# Horizontal
plt.figure(figsize=(8, 5))
plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.xlabel("Value")
plt.ylabel("Category")
plt.show()
```

```python
# Histograms

data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=20)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box plots
data1 = np.random.normal(0, 1, 200)
data2 = np.random.normal(2, 1.5, 200)

plt.figure(figsize=(8, 5))
plt.boxplot([data1, data2], labels=["Group 1", "Group 2"])
plt.title("Box Plot")
plt.ylabel("Value")
plt.show()
```

```python
# Pie charts
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [35, 20, 15, 30]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie Chart")
plt.show()
```

```python
# Subplots

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0, 0].plot(x, np.sin(x))
ax[0, 0].set_title("sin(x)")

ax[0, 1].plot(x, np.cos(x))
ax[0, 1].set_title("cos(x)")

ax[1, 0].plot(x, np.tan(x))
ax[1, 0].set_title("tan(x)")
ax[1, 0].set_ylim(-5, 5)

ax[1, 1].hist(np.random.randn(500), bins=20)
ax[1, 1].set_title("Historgram")

plt.tight_layout()
plt.show()

```

```python
# Titles, Labels, Legends, and Grid

x = np.linspace(0, 5, 100)

plt.figure(figsize=(8, 5))
plt.plot(x, x, label="x")
plt.plot(x, x**2, label="x^2")
plt.title("Labels and Legend Example")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Axis Limits and Tricks

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.xlim(0, 10)
plt.ylim(-1.2, 1.2)
plt.xticks(np.arange(0, 11, 2))
plt.yticks(np.arange(-1.0, 1.1, 0.5))
plt.title("Axis Limits and Ticks")
plt.show()
```

```python
# Markers, Line Styles, and Alpha

x = np.arange(1, 8)

plt.figure(figsize=(8, 5))
plt.plot(x, x, marker="o", linestyle="-", alpha=0.8, label="solid")
plt.plot(x, x+1, marker="s", linestyle="--", alpha=0.8, label="dashed")
plt.plot(x, x+2, marker="^", linestyle=":", alpha=0.8, label="dotted")
plt.legend()
plt.title("Markers and Line Styles")
plt.show()
```

```python
# Annotations and Text
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
peak_x = np.pi / 2
peak_y = np.sin(peak_x)

plt.annotate(
    "Peak",
    xy=(peak_x, peak_y),
    xytext=(peak_x + 1, peak_y + 0.2),
    arrowprops=dict(arrowstyle="->")
)
plt.text(7, -0.8, "Sine wave example")
plt.title("Annotations and Text")
plt.show()
```

```python
# Error Bars

x = np.arange(5)
y = np.array([2, 3, 5, 4, 6])
errors = np.array([0.2, 0.3, 0.4, 0.2, 0.5])

plt.figure(figsize=(8, 5))
plt.errorbar(x, y, yerr=errors, fmt="o-", capsize=5)
plt.title("Error Bar Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

```python
# Filling Areas
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.fill_between(x, y, alpha=0.3)
plt.title("Filled Area Under Curve")
plt.show()
```

```python
# Images with imshow
image = np.random.rand(20, 20)

plt.figure(figsize=(6, 6))
plt.imshow(image, cmap="viridis")
plt.colorbar()
plt.title("imshow Example")
plt.show()
```

```python
# Common Axes Methods
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot([1, 2, 3], [4, 5, 6])
ax.set_title("Axes Method Example")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True)
ax.legend(["line"])
plt.show()

common_axes_methods = [
    "plot", "scatter", "bar", "barh", "hist", "boxplot", "pie", "imshow",
    "set_title", "set_xlabel", "set_ylabel", "legend", "grid",
    "set_xlim", "set_ylim", "set_xticks", "set_yticks",
    "annotate", "text", "fill_between"
]

print("Common Axes methods:")
for method in common_axes_methods:
    print("-", method)
```

```python
print("Available styles:")
print(plt.style.available)

styles = plt.style.available
n = len(styles)

cols = 4
rows = (n + cols - 1) // cols

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, axes = plt.subplots(rows, cols, figsize=(16, rows * 3))
axes = axes.flatten()

for i, style in enumerate(styles):
    plt.style.use(style)
    
    ax = axes[i]
    ax.plot(x, y)
    ax.set_title(style, fontsize=9)
    
# Hide empty plots
for j in range(i + 1, len(axes)):
    axes[j].axis("off")

plt.tight_layout()
plt.show()
```

```python
print("Some public pyplot names:")

public_pyplot_names = [name for name in dir(plt) if not name.startswith("_")]

# limit to first 200
names = public_pyplot_names[:200]

cols = 5
rows = (len(names) + cols - 1) // cols

for r in range(rows):
    row_items = []
    for c in range(cols):
        idx = r + rows * c
        if idx < len(names):
            row_items.append(f"{names[idx]:<20}")
    print(" ".join(row_items))

print("\nTotal shown:", len(names))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
rng = np.random.default_rng(42)

fig, axes = plt.subplots(3, 3, figsize=(12, 10))

# 1. plot
axes[0, 0].plot(x, y)
axes[0, 0].set_title("plt.plot()")

# 2. scatter
axes[0, 1].scatter(x, y)
axes[0, 1].set_title("plt.scatter()")

# 3. bar
cats = ["A", "B", "C", "D"]
vals = [3, 7, 5, 6]
axes[0, 2].bar(cats, vals)
axes[0, 2].set_title("plt.bar()")

# 4. hist
data = rng.normal(0, 1, 500)
axes[1, 0].hist(data, bins=20)
axes[1, 0].set_title("plt.hist()")

# 5. pie
axes[1, 1].pie([25, 35, 20, 20], labels=["A", "B", "C", "D"], autopct="%1.1f%%")
axes[1, 1].set_title("plt.pie()")

# 6. errorbar
y_err = 0.2 + 0.2 * rng.random(10)
x2 = np.arange(10)
y2 = rng.random(10)
axes[1, 2].errorbar(x2, y2, yerr=y_err, fmt="o-")
axes[1, 2].set_title("plt.errorbar()")

# 7. stem
x3 = np.arange(10)
y3 = np.array([1, 3, 2, 5, 4, 6, 3, 2, 1, 4])
axes[2, 0].stem(x3, y3)
axes[2, 0].set_title("plt.stem()")

# 8. imshow
img = rng.random((10, 10))
axes[2, 1].imshow(img)
axes[2, 1].set_title("plt.imshow()")

# 9. contour
X, Y = np.meshgrid(np.linspace(-3, 3, 50), np.linspace(-3, 3, 50))
Z = np.sin(X**2 + Y**2)
axes[2, 2].contour(X, Y, Z)
axes[2, 2].set_title("plt.contour()")

plt.tight_layout()
plt.show()
```
