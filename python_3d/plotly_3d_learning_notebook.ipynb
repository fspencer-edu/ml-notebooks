## Imports

````python
import plotly.graph_objects as go
import numpy as np
````

## 3D scatter

````python
np.random.seed(42)
x, y, z = np.random.randn(3, 200)
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode="markers", marker=dict(size=4))])
fig.update_layout(title="3D Scatter")
fig.show()
````

## 3D line and surface

````python
t = np.linspace(0, 8*np.pi, 500)
fig = go.Figure(data=[go.Scatter3d(x=np.cos(t), y=np.sin(t), z=t, mode="lines")])
fig.show()

x = np.linspace(-5, 5, 80)
y = np.linspace(-5, 5, 80)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z)])
fig.show()
````
