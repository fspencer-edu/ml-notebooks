## Imports

````python
from mayavi import mlab
import numpy as np
````

## 3D points and surface

````python
x, y, z = np.random.random((3, 100))
mlab.points3d(x, y, z, scale_factor=0.05)
mlab.show()

x, y = np.mgrid[-3:3:100j, -3:3:100j]
z = np.sin(x**2 + y**2)
mlab.surf(x, y, z)
mlab.show()
````
