## Imports

````python
import pyvista as pv
import numpy as np
print("PyVista version:", pv.__version__)
````

## Meshes and plotting

````python
sphere = pv.Sphere()
cube = pv.Cube()
print(sphere)
plotter = pv.Plotter()
plotter.add_mesh(sphere, show_edges=True)
plotter.show()
````

## Point cloud and scalars

````python
points = np.random.rand(200, 3)
cloud = pv.PolyData(points)
plotter = pv.Plotter()
plotter.add_points(cloud, point_size=8, render_points_as_spheres=True)
plotter.show()

mesh = pv.Sphere(theta_resolution=30, phi_resolution=30)
mesh["z"] = mesh.points[:, 2]
plotter = pv.Plotter()
plotter.add_mesh(mesh, scalars="z")
plotter.show()
````
