````python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

print("Matplotlib imported successfully.")
````

## 2) Basic plotting with pyplot

````python

x = np.arange(0, 10)
y = x ** 2

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.show()
````

## 3) The Figure and Axes objects

````python

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y)
ax.set_title("Sine Wave")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
plt.show()

print("Figure object:", fig)
print("Axes object:", ax)
````

## 4) Common line plot options

````python

x = np.linspace(0, 10, 50)

plt.figure(figsize=(8, 5))
plt.plot(x, x, label="linear", linestyle="-", marker="o")
plt.plot(x, x**2, label="quadratic", linestyle="--", marker="s")
plt.plot(x, x**3, label="cubic", linestyle=":", marker="^")
plt.title("Multiple Lines")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
````

## 5) Scatter plots

````python

np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha=0.7)
plt.title("Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

## 6) Bar charts

````python

categories = ["A", "B", "C", "D"]
values = [10, 25, 15, 30]

plt.figure(figsize=(8, 5))
plt.bar(categories, values)
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
````

````python

plt.figure(figsize=(8, 5))
plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.xlabel("Value")
plt.ylabel("Category")
plt.show()
````

## 7) Histograms

````python

data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=20)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
````

## 8) Box plots

````python

data1 = np.random.normal(0, 1, 200)
data2 = np.random.normal(2, 1.5, 200)

plt.figure(figsize=(8, 5))
plt.boxplot([data1, data2], labels=["Group 1", "Group 2"])
plt.title("Box Plot")
plt.ylabel("Value")
plt.show()
````

## 9) Pie charts

````python

labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [35, 20, 15, 30]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie Chart")
plt.show()
````

## 10) Subplots

````python

x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("sin(x)")

axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title("cos(x)")

axes[1, 0].plot(x, np.tan(x))
axes[1, 0].set_title("tan(x)")
axes[1, 0].set_ylim(-5, 5)

axes[1, 1].hist(np.random.randn(500), bins=20)
axes[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
````

## 11) Titles, labels, legends, and grid

````python

x = np.linspace(0, 5, 100)

plt.figure(figsize=(8, 5))
plt.plot(x, x, label="x")
plt.plot(x, x**2, label="x^2")
plt.title("Labels and Legend Example")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
````

## 12) Axis limits and ticks

````python

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.xlim(0, 10)
plt.ylim(-1.2, 1.2)
plt.xticks(np.arange(0, 11, 2))
plt.yticks(np.arange(-1.0, 1.1, 0.5))
plt.title("Axis Limits and Ticks")
plt.show()
````

## 13) Markers, line styles, and alpha

````python

x = np.arange(1, 8)

plt.figure(figsize=(8, 5))
plt.plot(x, x, marker="o", linestyle="-", alpha=0.8, label="solid")
plt.plot(x, x+1, marker="s", linestyle="--", alpha=0.8, label="dashed")
plt.plot(x, x+2, marker="^", linestyle=":", alpha=0.8, label="dotted")
plt.legend()
plt.title("Markers and Line Styles")
plt.show()
````

## 14) Annotations and text

````python

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
peak_x = np.pi / 2
peak_y = np.sin(peak_x)

plt.annotate(
    "Peak",
    xy=(peak_x, peak_y),
    xytext=(peak_x + 1, peak_y + 0.2),
    arrowprops=dict(arrowstyle="->")
)
plt.text(7, -0.8, "Sine wave example")
plt.title("Annotations and Text")
plt.show()
````

## 15) Error bars

````python

x = np.arange(5)
y = np.array([2, 3, 5, 4, 6])
errors = np.array([0.2, 0.3, 0.4, 0.2, 0.5])

plt.figure(figsize=(8, 5))
plt.errorbar(x, y, yerr=errors, fmt="o-", capsize=5)
plt.title("Error Bar Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
````

## 16) Filling areas

````python

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.fill_between(x, y, alpha=0.3)
plt.title("Filled Area Under Curve")
plt.show()
````

## 17) Images with imshow

````python

image = np.random.rand(20, 20)

plt.figure(figsize=(6, 6))
plt.imshow(image, cmap="viridis")
plt.colorbar()
plt.title("imshow Example")
plt.show()
````

## 18) Working with pandas data

````python

df = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "sales": [120, 150, 170, 160, 190],
    "profit": [20, 30, 35, 28, 40]
})

plt.figure(figsize=(8, 5))
plt.plot(df["month"], df["sales"], marker="o", label="Sales")
plt.plot(df["month"], df["profit"], marker="s", label="Profit")
plt.title("Sales and Profit by Month")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.legend()
plt.show()

df
````

## 19) Saving figures

````python

x = np.linspace(0, 10, 100)
y = np.cos(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Figure Saved to File")
plt.savefig("matplotlib_example_plot.png", dpi=150, bbox_inches="tight")
plt.show()

print("Saved figure as matplotlib_example_plot.png")
````

## 20) Common Axes methods

````python

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot([1, 2, 3], [4, 5, 6])
ax.set_title("Axes Method Example")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True)
ax.legend(["line"])
plt.show()

common_axes_methods = [
    "plot", "scatter", "bar", "barh", "hist", "boxplot", "pie", "imshow",
    "set_title", "set_xlabel", "set_ylabel", "legend", "grid",
    "set_xlim", "set_ylim", "set_xticks", "set_yticks",
    "annotate", "text", "fill_between"
]

print("Common Axes methods:")
for method in common_axes_methods:
    print("-", method)
````

## 21) Common pyplot functions

````python

common_pyplot_functions = [
    "figure", "subplots", "plot", "scatter", "bar", "barh", "hist",
    "boxplot", "pie", "imshow", "title", "xlabel", "ylabel", "legend",
    "grid", "xlim", "ylim", "xticks", "yticks", "annotate", "text",
    "tight_layout", "savefig", "show", "colorbar", "errorbar", "fill_between"
]

print("Common pyplot functions:")
for func in common_pyplot_functions:
    print("-", func)
````

````python

plt.style.use("ggplot")

x = np.linspace(0, 10, 100)
plt.figure(figsize=(8, 5))
plt.plot(x, np.sin(x), label="sin(x)")
plt.title("Using a Matplotlib Style")
plt.legend()
plt.show()

# Reset to default if you want:
plt.style.use("default")
````

## 23) Layout helpers

````python

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(np.arange(10), np.arange(10))
axes[0].set_title("Left")

axes[1].plot(np.arange(10), np.arange(10)**2)
axes[1].set_title("Right")

plt.tight_layout()
plt.show()
````
